```python
#Opening a csv with url

import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

```python
#index_col parameter
pd.read_csv("/content/aug_train.csv",index_col="enrollee_id")
```

```python
#header parameter
pd.read_csv("/content/aug_train.csv",header=1)
```

```python
#usecols parameter
pd.read_csv("/content/aug_train.csv",usecols=["enrollee_id","gender","education_level"])
```

```python
#for skipping rows
pd.read_csv("/content/aug_train.csv",skiprows=[0,1])
```

```python
#for limiting data to n rows
pd.read_csv("/content/aug_train.csv",nrows=150)
```

```python
#dtype parameter
pd.read_csv("/content/aug_train.csv",dtype={"target":int}).info()
```

```python
#Handling dates
pd.read_csv("/content/aug_train.csv",parse_dates=['cols name ehose dtype you wanna change as date'])
```
